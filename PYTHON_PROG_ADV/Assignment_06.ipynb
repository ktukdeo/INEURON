# 1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

```python
def find_the_difference(str1, str2):
    lst1=list(str1);lst2=list(str2)
    lst1.sort()
    lst2.sort()
    for i in range(len(lst1)):
        if lst1[i] != lst2[i]:
            return lst2[i]
        
    return lst2[-1]

print('Diff: ',find_the_difference("abcd", "abcde") )
print('Diff: ',find_the_difference("", "y") )
print('Diff: ',find_the_difference("ae", "aea") )
```

# 2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.

```python
lst_data = [int,str,bool,list,tuple,dict]
def count_datatypes(*arg):
    lst = [0] * 6
    for x in arg:
        i = lst_data.index(type(x))
        lst[i] += 1
        
    print(arg,'   ',lst)
    
count_datatypes(1, 45, "Hi", False) 
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) 
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23) 
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

# 3. A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) 

```python
def fib_str(seq,lst):
    hist={1:lst[0], 2:lst[1]}
        
    def fib_seq(seq,hist):
        if seq in hist:
            return hist[seq], hist
        else:
            fs = fib_seq(seq-1,hist)[0] + fib_seq(seq-2,hist)[0]
            hist[seq] = fs
            return fs, hist
    
    _,h = fib_seq(seq,hist)
    return [v for v in h.values()]

print(fib_str(3, ["j", "h"]))
print(fib_str(5, ["e", "a"]))   
print(fib_str(6, ["n", "k"]))
```

# 4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:

```python
def ones_threes_nines(num):
    x9 = num//9; num -= x9 * 9
    x3 = num//3; num -= x3 * 3
    x1 = num
    
    return f"nines:{x9}, threes:{x3}, ones:{x1}"

print(ones_threes_nines(10) )
print(ones_threes_nines(15) )
print(ones_threes_nines(22) )
```

# 5. Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n.

```python
#Non-recursive
def fib(seq):
    if seq in [1,2]: return 1
    prev1=1;prev2=1
    for i in range(3,seq+1):
        prev1,prev2 = prev1+prev2,prev1
    return prev1

print(fib(6))
print(fib(10) )
```
